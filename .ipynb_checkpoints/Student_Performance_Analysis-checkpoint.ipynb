```python
df[['math score', 'reading score', 'writing score']].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.barplot(x='gender', y='math score', data=df)
plt.title("Average Math Score by Gender")
plt.show()
```

Students who completed the test preparation course tend to score higher.

```python
plt.figure(figsize=(6,4))
sns.barplot(x='test preparation course', y='math score', data=df)
plt.title("Effect of Test Preparation on Math Score")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(
    df[['math score', 'reading score', 'writing score']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Between Scores")
plt.show()
```

Strong positive correlation exists between math, reading, and writing scores.

# Student Performance Data Analysis

## Project Overview
This project analyzes student academic performance using data science techniques.
The main objective is to explore how different factors such as gender, test preparation,
and subject-wise scores influence overall student performance.

---

## Dataset Description
The dataset contains academic performance records of 1000 students with 8 attributes:
- Gender  
- Race/Ethnicity  
- Parental level of education  
- Lunch type  
- Test preparation course  
- Math score  
- Reading score  
- Writing score  

The dataset does not contain any missing values, making it suitable for exploratory data analysis.

---

## Exploratory Data Analysis (EDA)
Initial exploration shows that the dataset includes both categorical and numerical variables.
Descriptive statistics indicate that students generally perform better in reading and writing
compared to mathematics.

The average scores are:
- Math score: ~66  
- Reading score: ~69  
- Writing score: ~68  

This suggests that mathematics is comparatively more challenging for students.

---

## Visual Analysis and Insights
Several visualizations were created to understand performance patterns:

- Gender-wise analysis shows that male students have a slightly higher average math score than female students.
- Students who completed the test preparation course consistently scored higher in mathematics.
- Bar charts highlight clear performance differences based on preparation and demographic factors.

---

## Correlation Analysis
Correlation analysis reveals a strong positive relationship between math, reading, and writing scores.
In particular:
- Reading and writing scores show a very strong correlation.
- Math scores are also positively correlated with reading and writing scores.

This indicates that students who perform well in one subject are likely to perform well in others.

---

## Key Findings
- Test preparation courses have a positive impact on student performance.
- Reading and writing skills are closely related.
- Mathematics is the most challenging subject on average.
- Gender-based differences exist but are not significantly large.

---

## Conclusion
This project demonstrates how exploratory data analysis and visualization techniques
can be used to extract meaningful insights from educational data.
The findings can help educators and policymakers better understand factors that influence
student performance and support data-driven academic planning.
